### Null values

```python

df.isna().sum()
```

```python
age_mean=df["Age"].mean()
df["Age"]=df["Age"].fillna(age_mean)
df
```

```python
df=df.dropna(subset="Embarked")
```

### Duplicates

```python
#count duplicates
df.duplicated().sum()
```

```python
#remove duplicates
df.drop_duplicates()
```

### Unique values

```python
#Find unique values
df["Embarked"].unique()
```

```python
#cFind count of unique values
df["Embarked"].value_counts()
```

```python
#Rename coloumn
df.rename(columns={"Pclass":"Passenger_Class"},inplace=True)
df.head(1)
```

```python
#sort values
df.sort_values("Age",inplace=True)
df.head(5)
```

### Feature Extraction

```python
def age_group(age):
    if age<13:
        return "Child"
    if age<20:
        return "Teen"
    if age<60:
        return "Adult"
    else:
        return "Senior"
    
df["Age_Group"]=df["Age"].apply(age_group)
df.head()
```

```python
df["Age_Group"].value_counts()
```

### Grouping

```python
df.groupby("Age_Group")["Survived"].mean()*100
```

```python
df[["Age_Group","Survived"]].value_counts()
```

```python
df["Ticket"].value_counts().head(5)
```

### Pivot tables

```python
df.pivot_table(
    values="Survived",
    index="Sex",
    columns="Passenger_Class",
    aggfunc="mean"
)
```

```python
df.pivot_table(
    values="Fare",
    index="Passenger_Class",
    aggfunc="count"
)
```

### Encoding

```python
#Encoding
df["Sex"]=df["Sex"].map({
    "male":0,
    "female":1
})

df.head()
```

### Exporting 

```python
#Export Datset
df.to_csv("cleaned_titanic.csv",index=False)
```
